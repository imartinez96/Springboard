```python
AS_data = bb_data.groupby('Allstar')[['AVG','SLG', 'HR', 'RBI', 'OPS', 'OBP', 'XBH']].mean()
```

```python
AS_data.head()
```

```python
(AS_data.reindex(index=AS_data.mean(axis=1)
    .sort_values(ascending=False)
    .index)
    .plot(kind='barh', figsize= (10,10), title='Average stats by Allstar and Non-Allstar'))
plt.xlabel('Stats');
```
